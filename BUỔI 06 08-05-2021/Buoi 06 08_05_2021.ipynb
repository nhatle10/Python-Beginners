```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([[65, 69, 75, 76, 78, 80]]).T
Y = np.array([[170, 173, 175, 176, 177, 182]]).T
```

```python
# Visualize data 
plt.scatter(X, Y, c = 'red')
plt.xlabel('Weight (kg)')
plt.ylabel('Hight (cm)')
plt.show()
```

```python
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis = 1)
```

```python
print(one)
```

```python
print(Xbar)
```

```python
A = np.dot(Xbar.T, Xbar)
```

```python
print(A)
```

```python
B = np.dot(Xbar.T, Y)
```

```python
print(B)
```

```python
w = np.dot(np.linalg.pinv(A), B)
print('w = ', w)
# Preparing the fitting line 
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(145, 185, 2)
y0 = w_0 + w_1*x0
```
